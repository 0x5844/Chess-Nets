# cheeseNet 🧀 - v0.1-alpha

### This notebook showaces the setup to run 🧀 locally. It utilizes the following technologies:
1. Lichess Bot: [lichess-bot Python SDK](https://github.com/lichess-bot-devs/lichess-bot)
2. Engine: [Lc0 Engine](https://lczero.org/)

Here are the setup steps:

```python
# 1. Installed required packages
!apt-get update && apt-get -y install clang-6.0 ninja-build
!pip3 install meson==0.55.3

# 2. Clone Lc0 locally
!rm -rf lc0
!git clone --recurse-submodules https://github.com/LeelaChessZero/lc0.git
```

```python
# 3. Build lc0
!cd lc0 && rm -rf build
!chmod +x lc0/build.sh
!cd lc0 && ./build.sh
!chmod +x lc0/build/lc0
```

```python
# 4. Clone lichess-bot SDK locally
!git clone https://github.com/ClaustrophobicPie/lichess-bot

# 5. Setup lichess-bot
!cp lc0 lichess-bot/engines -r
!cp lichess-bot/config.yml.default lichess-bot/config.yml 
# Make sure you API key is set in config.yml before moving forward.
# You should also adjust the engine path (https://github.com/lichess-bot-devs/lichess-bot/wiki/Setup-the-engine)
# for Lc0 you can use https://github.com/ClaustrophobicPie/Chess-Nets/blob/main/networks/cheeseNet-T82-768x15x24h-swa-964000.pb.gz
```

```python
# 6. Setup lichess-bot environment
!cd lichess-bot/ && python3 -m venv venv
!cd lichess-bot/ && source venv/bin/activate
!cd lichess-bot/ && pip install -r requirements.txt
```

```python
# 7. Run lichess-bot
!cd lichess-bot/ && python3 lichess-bot.py
```
